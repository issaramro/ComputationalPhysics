```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# y' = 2y
# f = 2y
h = 0.005
upper_bound = 1
t = np.arange(0, upper_bound, h)
y = np.zeros(len(t))
y_0 = 1
y[0] = y_0
for i in range(0, int((upper_bound)/h)-1):
    y [i + 1] = y[i] + 2*h*y[i] 
y_a = np.exp(2*t)
plt.plot(t,y, label = "Euler method")
plt.plot(t,y_a, label = "analytical solution")
plt.legend()
plt.show()
l2_error = np.sum((y_a - y)**2)
print(l2_error/len(y))
```

```python
# x'' + w^2x = 0
# g = - w^2 x
h = 0.005
upper_bound = 10
t = np.arange(0, upper_bound, h)
x = np.zeros(len(t))
v = np.zeros(len(t))
w_2 = 4
x_0 = 1
v_0 = 0
x[0] = x_0
v[0] = v_0
for i in range(0, int((upper_bound)/h)-1):
     x [i + 1] = x[i] + h*v[i]
     v [i + 1] = v[i] - h*w_2*x[i]
x_a = np.cos(2*t)
plt.plot(t,x, label = "Euler method")
plt.plot(t,x_a, label = "analytical solution")
plt.legend()
plt.show()
l2_error = np.sum((x_a - x)**2)
print(l2_error/len(x))
```
