```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def julia(n_range, m_range, max, a):
    y, x = np.ogrid[1.4: -1.4: n_range*1j, -1.4: 1.4: m_range*1j]
    z_array = x + y*1j
    div_iterations = max + np.zeros(z_array.shape)
    for n in range(n_range):
        for m in range(m_range):
            z = z_array[n][m]
            for i in range(max):
                z = z**2 + a
                if z * np.conj(z) > 2:
                    div_iterations[n][m] = i
                    break
    return div_iterations

plt.imshow(julia(1000, 1000, 20, -0.8 + 0.2j), cmap='nipy_spectral')
plt.axis('off')
plt.show()
```
