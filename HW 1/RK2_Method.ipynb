```python
# u'' = - u + a
h = 0.001
upper_bound = 10
theta = np.arange(0, upper_bound, h)
u = np.zeros(len(theta))
v = np.zeros(len(theta))
u_0 = 1
v_0 = 0
a = 1.65389512 * 10**(-24)
u[0] = u_0
v[0] = v_0
for i in range(0, int(upper_bound/h)-1):
    k1 = h * v[i]
    k2 = -h * u[i] + h*a
    u[i+1] = u[i] + h * (v[i] + k2/2)
    v[i+1] = v[i] - h * u[i] + h* a
u_a = 0.9 *np.cos(theta) +  1.65389512 * 10**(-24)
plt.plot(theta, u_a, label="analytical solution")
plt.plot(theta, u, label="RK2 method")
plt.legend()
plt.show()
l2_error = np.sum((u_a - u)**2)/len(u)
print(l2_error)
```
