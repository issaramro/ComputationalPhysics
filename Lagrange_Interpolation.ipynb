```python
import numpy as np
import matplotlib.pyplot as plt
```

# First order Lagrange interpolation:

```python
x = np.linspace(-25,25, 10)
noise = 1.5*np.random.randint(-10, 10)
y = x**2 + noise
y1 = x**2
plt.scatter(x, y, label = "y = x^2 + noise")
plt.scatter(x, y1, label = "y = x^2")
for i in range(0, len(x)-1):
    x1 = x[i]
    x2 = x[i+1]
    
    y1 = y[i]
    y2 = y[i+1]


    j = 1 # order
    i = np.linspace(x1,x2)
    p = y1 * ((i - x2)/(x1 - x2)) + y2 *((i-x1)/(x2-x1))
    plt.plot(i,p)
plt.ylim(-100, 800)
plt.legend()
plt.show()
```

# Second order Lagrange interpolation:

```python
x = np.linspace(-25,25, 10)
noise = 1.5*np.random.randint(-10, 10)
y = x**2 + noise
y1 = x**2
plt.scatter(x, y, label = "y = x^2 + noise")
plt.scatter(x, y1, label = "y = x^2")
for i in range(0, len(x)-2):
    x1 = x[i]
    x2 = x[i+1]
    x3 = x[i+2]
    
    y1 = y[i]
    y2 = y[i+1] 
    y3 = y[i+2]

    j = 2 # order
    i = np.linspace(x1, x3)
    p = y1 * ((i - x2)/(x1 - x2))*((i-x3)/(x1-x3)) + y2 *((i-x1)/(x2-x1))*((i-x3)/(x2-x3)) + y3*((i-x2)/(x3-x2))*((i-x1)/(x3-x1))
    plt.plot(i,p)
plt.ylim(-100,800)
plt.legend()
plt.show()
```


# Now, defining the order by the nb of known points in my interval:

```python
x = np.linspace(-100,100, 10)
y = x**2 + 1.5*np.random.randint(-20,20)
y1 = x**2
j = len(x) - 1 # order

for i in np.arange(x[0], x[len(x)-1] + 1, 5):
    x_given = i
    
    y_we_seek = 0

    for i in range(j+1):
        p =1
        for j in range(j+1):
            if j != i:
                p *= ((x_given - x[j])/(x[i] - x[j]))
        y_we_seek += y[i]*p

    plt.scatter(x_given, y_we_seek, color = "r")

plt.scatter(x,y, label = "y = x^2 + noise")
plt.scatter(x,y1, label = "y = x^2")
plt.legend()
plt.show()
```

# Now ploting the function instead of scattered points:

```python
x = np.linspace(-100,100, 10)
y = x**2 + 1.5*np.random.randint(-20,20)
y1 = x**2
j = len(x) - 1 # order
x_given_int = []
y_we_seek_int = []

for i in np.arange(x[0], x[len(x)-1] + 1, 2):
    x_given = i
    x_given_int.append(x_given)

    y_we_seek = 0


    for i in range(j+1):
        p =1
        for j in range(j+1):
            if j != i:
                p *= ((x_given - x[j])/(x[i] - x[j]))
        y_we_seek += y[i]*p
    y_we_seek_int.append(y_we_seek)


plt.plot(x_given_int, y_we_seek_int, color = "r", label = "the interpolated function")


plt.scatter(x,y, label = "y = x^2 + noise")
plt.scatter(x,y1, label = "y = x^2")
plt.legend()
plt.show()
```
