```python
import numpy as np
import matplotlib.pyplot as plt

def Newton_Method(f, df,a,b, p, k):
    r= np.zeros(k)
    root = 0
    # 2 bisections first
    r[0] = (a+b)/2
    if f(a)*f(r[0]) < 0:
        r[1] = (a + r[0])/2
    elif f(b)*f(r[0]) < 0:
        r[1]= (r[0] + b)/2
    else:
        root = r[0]
        return
    n = 1
    tol = 0.5 * 10**(-p + 1)
    error = tol + 1
    while (error > tol and n <k-1):
       r[n+1] = r[n] - f(r[n])/df(r[n])
       error = np.abs((r[n+1] - r[n])/ r[n+1])
       n = n+1
    if n > k:
        print("no convergence")
    else:
        root = r[len(r) - 1]
    return root

def f(x):
    return x*np.exp(x) - 5*np.exp(x) + 5
def df(x):
    return np.exp(x) + x**2 * np.exp(x) - 5*x*np.exp(x)

x = np.arange(-10, 10)
plt.plot(x, f(x))
plt.show()
print(Newton_Method(f, df, 4.5, 5, 10, 10))
```
