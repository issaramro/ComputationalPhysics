```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, u, k, alpha, a):
    return (k * alpha * x) / (np.sqrt(np.abs(alpha*x)**2 + (a - u)**2) - k * (a - u))


def rk4(f, x0, y0, xf, k, alpha, a,h):
    #h = 0.001
    x = np.arange(x0, xf, h)
    y = np.zeros(len(x))
    y[0] = y0
    for i in range(0,len(x) - 1):
        k1 = h * f(x[i], y[i], k, alpha, a)
        k2 = h * f(x[i] + 0.5*h, y[i] + 0.5*k1, k, alpha, a)
        k3 = h * f(x[i] + 0.5*h, y[i] + 0.5*k2, k, alpha, a)
        k4 = h * f(x[i] + h, y[i] + k3, k, alpha, a)

        y[i+1] = y[i] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return x, y

x0 = 0
xf = 100
a = 500
u0 = 70
k = 1.5168
alpha = [-2,-0.5, 0.5, 2]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 15))    
x1, y1 = rk4(f, 0, u0, 100, k,-2, a,0.001)
x1_, y1_ = rk4(f, 0, u0, -100, k, -2, a,-0.001)
ax1.plot(x1,y1, color = 'b')
ax1.plot(x1_, y1_, color = 'b')
ax1.set_title(" = - 2")
ax1.set_xlabel('x')
ax1.set_ylabel('(x)')
ax1.grid(True)

x2, y2 = rk4(f, 0, u0, 100, k,-0.5, a, 0.001)
x2_, y2_ = rk4(f, 0, u0, -100, k, -0.5, a,-0.001)
ax2.plot(x2,y2,color = 'b')
ax2.plot(x2_, y2_,  color = 'b')
ax2.set_title(" = - 0.5")
ax2.set_xlabel('x')
ax2.set_ylabel('(x)')
ax2.grid(True)

x3, y3 = rk4(f, 0, u0, 100, k,2, a,0.001)
x3_, y3_ = rk4(f, 0, u0, -100, k, 2, a,-0.001)
ax3.plot(x3,y3,color = 'b')
ax3.plot(x3_, y3_,  color = 'b')
ax3.set_title(" = 2")
ax3.set_xlabel('x')
ax3.set_ylabel('(x)')
ax3.grid(True)

x4, y4 = rk4(f, 0, u0, 100, k,0.5, a,0.001)
x4_, y4_ = rk4(f, 0, u0, -100, k, 0.5, a,-0.001)
ax4.plot(x4,y4, color = 'b')
ax4.plot(x4_, y4_, color = 'b')
ax4.set_title(" = 0.5")
ax4.set_xlabel('x')
ax4.set_ylabel('(x)')
ax4.grid(True)

# fig.suptitle("(0) = 70, a = 500, k = 1.5168")
plt.show()
```
