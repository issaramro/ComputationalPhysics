```python
import numpy as np
import matplotlib.pyplot as plt

def psi(r, theta, u, a, w):
    return np.sin(theta) * u * (r - a**2/r) + (w/(2 * np.pi)) * theta

r = np.linspace(1, 1000, 100)
theta = np.linspace(0, 2 * np.pi, 100)
u = 5
a = 10
w = 1

x = r * np.cos(theta)
y = r * np.sin(theta)

psi_values = psi(np.sqrt(x**2 + y**2), np.arctan(y/x), u, a, w)
ux = np.gradient(psi_values, x)
uy = -np.gradient(psi_values, y)

plt.figure(figsize=(8, 8))
plt.plot(10*np.cos(theta), 10*np.sin(theta), color='r')
plt.quiver(x,y,ux,uy, scale=20, color='b')  # Adjust the scale as needed
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quiver Plot')
plt.show()
```
