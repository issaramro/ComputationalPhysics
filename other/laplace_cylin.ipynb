```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.linspace(1,100, 100)
def phi(r, theta, R, u):
    return (r + (R**2)/r)*u*np.cos(theta)

theta, r = np.meshgrid(t,x)
p = phi(r, theta, 5, 10)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(r, theta, p, cmap='viridis')

ax.set_xlabel('r')
ax.set_ylabel('θ')
ax.set_zlabel('Φ(r,θ)')

plt.show()
```
